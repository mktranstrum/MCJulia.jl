```python
import corner
import numpy as np
```

```python
filename = "./chains/line_fit_chain.npy"
label_list = [
        r"$a$",
        r"$b$",
        r"$\sigma$",
    ]

# If doing Gaussian fit example, use different label list and filename = "./chains/gaussian_fit_chain.npy"

samples = np.load(filename)

# Julia and python store data column-wise and row-wise. Need to transpose for use with corner.py
samples = samples.T
```

```python
figure = corner.corner(
    samples,
    labels=label_list,
    quantiles=[0.16, 0.5, 0.84],
    show_titles=True,
    title_kwargs={"fontsize": 12},
)
```
